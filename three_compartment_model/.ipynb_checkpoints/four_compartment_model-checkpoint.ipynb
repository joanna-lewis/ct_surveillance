# Chlamydia testing and prevalence

The four-compartment model is designed to try to understand the relationship between current chlamydia infection and testing in the last year. In Natsal-3, 66.4% of those with a chlamydia-positive urine test had not been tested in the last year. 


```python
from IPython.display import Image
Image(filename="figures/4_comp.png", width=300)
```

Compartments in the model correspond to those who are infected vs. uninfected, and tested vs. untested. At the "beginning" of a year, the population is split between the two untested compartments in steady-state proportions determined by the incidence and natural clearance rates.

From this initial condition, we integrate over one year to obtain the proportion of the population in each of the four compartments after one year.

We assume natural clearance occurs at a rate 0.74 (Price _et al._, Stat. Med. 2013), and vary incidence and testing rate over plausible ranges.

```python
clear = 0.74 # set rate of natural chlamydia clearance
```

```python
from scipy.integrate import odeint

%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt

inc = linspace(0.001, 0.1, 101) # incidence
test = linspace(0.01, 1, 101) # test
inc,test = meshgrid(inc, test)

P = empty([101, 101, 4]) # (P_11, P_21, P_12, P_22)

for i in xrange(101):
    for j in xrange(101):
        init = [clear/(inc[i,j] + clear), inc[i,j]/(inc[i,j] + clear), 0, 0]
        def fun(y, t):
            dP_11 = -(inc[i,j]+test[i,j])*y[0] + clear*y[2]
            dP_21 = -inc[i,j]*y[1] + test[i,j]*(y[0]+y[2]) + clear*y[3]
            dP_12 = inc[i,j]*y[0] - (clear + test[i,j])*y[2]
            dP_22 = inc[i,j]*y[1] - clear*y[3]
            return([dP_11, dP_21, dP_12, dP_22])
            
        sol = odeint(fun, init, [0,1])
        P[i,j] = sol[1]
```

```python
fig = plt.figure(figsize = (12, 7))

ax1 = fig.add_subplot(221)
p = ax1.pcolor(inc,test, P[:,:,0])
cb = fig.colorbar(p, ax=ax1)
ax1.set_ylabel('Testing Rate')
t = ax1.text(0.05, 0.9, 'Uninfected, untested', ha='center', size='large')
t.set_bbox(dict(facecolor='white', alpha=0.7, edgecolor='None'))
ax1.set_ylim(0, 1)
ax1.set_xlim(0, 0.1)

ax2 = fig.add_subplot(222)
p = ax2.pcolor(inc,test, P[:,:,2])
cb = fig.colorbar(p, ax=ax2)
t = ax2.text(0.05, 0.9, 'Infected, untested', ha='center', size='large')
t.set_bbox(dict(facecolor='white', alpha=0.7, edgecolor='None'))
ax2.set_ylim(0, 1)
ax2.set_xlim(0, 0.1)

ax3 = fig.add_subplot(223)
p = ax3.pcolor(inc,test, P[:,:,1])
cb = fig.colorbar(p, ax=ax3)
t = ax3.text(0.05, 0.9, 'Uninfected, tested', ha='center', size='large')
t.set_bbox(dict(facecolor='white', alpha=0.7, edgecolor='None'))
ax3.set_xlabel('Incidence')
ax3.set_ylabel('Testing Rate')
ax3.set_ylim(0, 1)
ax3.set_xlim(0, 0.1)

ax4 = fig.add_subplot(224)
p = ax4.pcolor(inc,test, P[:,:,3])
cb = fig.colorbar(p, ax=ax4)
ax4.set_xlabel('Incidence')
t = ax4.text(0.05, 0.9, 'Infected, tested', ha='center', size='large')
t.set_bbox(dict(facecolor='white', alpha=0.7, edgecolor='None'))
ax4.set_ylim(0, 1)
ax4.set_xlim(0, 0.1)

plt.show()   
```

We can now calculate the proportion of those infected who have not been tested during the previous year, and compare it to the Natsal-3 estimate and 95% confidence interval of 66.4% (55.1% - 76.1%).

```python
p = plt.pcolor(inc,test, P[:,:,2]/(P[:,:,2] + P[:,:,3]))
c = plt.contour(inc,test, P[:,:,2]/(P[:,:,2] + P[:,:,3]), [0.551, 0.664, 0.761], colors='k')
plt.clabel(c)
plt.colorbar(p)
plt.xlabel('Incidence')
plt.ylabel('Testing Rate')
plt.title('P(not tested in the last year | currently infected)')
```
